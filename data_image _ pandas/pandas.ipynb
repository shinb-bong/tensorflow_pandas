```python
import pandas as pd

raw = pd.read_excel('product.xlsx',engine="openpyxl")

raw['부가세포함'] = raw['판매가'] * 1.1

def 함수(n) :
  n = n *1.1
  return n

raw['실제가격']= raw['판매가'].apply(함수) # 특정 컬럼에 있는 모든 데이터를 함수에 넣었다가 뺴줌
raw

import re # 정규식라이브러리
re.search('abc','abcdef')
def 카테고리함수(n):
  if re.search('Chair',str(n)):
    return "Chair"
  if re.search('Table',str(n)):
    return "Table"


raw['카테고리'] = raw['상품목록'].apply(카테고리함수)
raw
```

```python
# 정규식( 문자를 검사하는 식 )

import re

re.search('정규식','안녕하세요 정규식') # 있으면 Object 자료형 뱉어줌 없으면 None

a = re.findall('\$','abcde$fg') # findall있으면 전부 리스트에 담아줌 ^시작문자 ,  끝나는문자$ , [이거저거]
a = re.findall('[가-힣ㄱ-ㅎ]','abAcde$fg하이ㅎㅎ') # [이거저거] 이거 or 저거 찾는거 [] 안에서 ^쓰면 not 임 \d하면 모든 숫자 \d\d 두자리 숫자 =  \d{2}  숫자가 아닌건 \D / \s 스페이스바 찾고 싶으면 \S 는 모든 문자
a = re.findall('[ㅋ+ abA]','abAcde$fg하이ㅎㅎㅋㅋㅋㅋㅋ',re.IGNORECASE) # 'ㅋ+' ㅋ 붙어있는거 다가지고옴 
# '2022-1-1'
b = re.sub('\-','.','2022-1-1') # .sub() =>이걸찾아서, 이걸로바꿔, 문장
b = re.sub('\d','','202020.20.20')
print(b)
print(a)
```
