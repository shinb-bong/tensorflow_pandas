Q1. 아마존 (종목코드 AMZN) 주식의 2020년 01월01일~12월31일 주식가격을 yahoo에서 가져온 후 20일, 60일 이동평균선을 그려보십시오. 

```python
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt

df = data.DataReader('AMZN','yahoo',start='2020-01-01', end='2020-12-31')
plt.plot(df.index,df['Close'].rolling(20).mean())
plt.plot(df.index,df['Close'].rolling(60).mean())
plt.show()
```

Q2. 삼성전자와 경쟁업체 LG전자의 매출 데이터는 다음과 같습니다.  

Samsung	LG
2018	50000	30000
2019	60000	40000
2020	75000	50000
2021	70000	35000
 

팀장님이 이걸 이쁜 그래프로 만들어서 보고를 올리라고 하는군요.

그러니 matplotlib을 이용해 이쁘게 그려보십시오. 팀장님은 연세가 있어 x,y축 라벨과 legend도 친절하게 적어드려야합니다. 

```python
import matplotlib.pyplot as plt

plt.plot([2018,2019,2020,2021],[50000,60000,75000,70000])
plt.plot([2018,2019,2020,2021],[30000,40000,50000,35000])
plt.xlabel('Time')
plt.ylabel('Sale')
plt.legend(['Samsung','LG'])
plt.show()
```

Q3. 원하는 데이터만 보이게 그래프를 조작하고 싶습니다. 

yahoo에서 2020년 01월01일~12월31일의 비트코인 가격을 가져오고 Close 가격을 그래프로 그리고 싶습니다. (종목코드 BTC-USD)

근데 Volume항목 2020년의 평균Volume보다 높은 날의 가격만 그래프로 그려보고 싶은데 어떻게 코드를 짜야할까요? 

```python
import pandas as pd
from pandas_datareader import data

df = data.DataReader('BTC-USD','yahoo', start="2020-01-01", end="2020-12-31")
avg = df['Volume'].mean()
print(avg)

df['tf'] = df['Volume'] > avg
print(df)

a = df[df['tf']==True]  # 새로운 a라는 데이터 프레임 만들고 가져오기
print(a)


import matplotlib.pyplot as plt
plt.bar(a.index, a['Close'])
plt.show() 
```
